```python
from transformers import pipeline
```

# Configuração do Gerador

```python
gerador = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B')
```

# Texto gerado pelo prompt

```python
prompt = "What is the thermodynamics?"
```

```python
resposta = gerador(prompt, max_length=100, min_length=30, do_sample=false, temperature=0.9)
```

# Saída do texto

```python
print(resposta[0]['generated_text'])
```

# Exportar texto

```python
with open('gpgtexto.txt', 'w') as f:
  f.writelines(resposta[0]['generated_text'])
```
